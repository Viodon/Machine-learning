```python
#1.	Виведіть перші 5 рядків таблиці.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

# Зчитуємо таблицю в DataFrame
data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця

df_1 = data.head(5)
# Виводимо перші 5 рядків таблиці
print(df_1)
```

```python
#2.	Визначте кількість рядків і стовпчиків за допомогою атрибуту shape

numRows, numColumns = data.shape
print(f"Кількість рядків: {numRows}, Кількість стовпчиків: {numColumns}")
```

```python
#3.	Визначте загальну кількість років та кількість місяців у даних.

numYears = len(data['Рік']) - 1  # віднімання 1, оскільки перший стовпець - назва
numMonths = numYears*12
print("Загальна кількість років =", numYears)
print("Загальна кількість місяців =", numMonths)
```

```python
#4.	Замінити пропущені значення  «-» в таблиці на NaN

data = data.rename(columns={
    "Рік": "Year",
    "В національній валюті": "Salary in UAH",
    "В доларах США": "Salary in USD",
    "Кількість кг хліба[20] на середню зарплату": "Bread per Salary",
    "Кількість БігМаків на середню зарплату": "BigMacs per Salary"
})
data = data.replace("—", np.nan)
data
```

```python
#5.	Визначити тип стовпців за допомогою dtypes.

columnTypes = data.dtypes
print("Типи стовпців:")
print(columnTypes)
```

```python
#6.	Видаліть передостанній і останній стовпчики ( Кількість кгхліба[20] на середнюзарплату,  Кільксть БігМаків на середню зарплату). Далі працюйте з оновленою таблицею.

data = data.drop(columns=["Bread per Salary", "BigMacs per Salary"])
data
```

```python
#7.	Замінити тип нечислових даних на числові.

data['Salary in UAH'] = data['Salary in UAH'].str.replace('▲', '').str.replace('▼', '').str.replace(',', '.').astype(float)
data['Salary in USD'] = data['Salary in USD'].str.replace('▲', '').str.replace('▼', '').str.replace(',', '.').astype(float)
data["Year"] = data["Year"].str.replace("*", "").astype(int)

data.dtypes
```

```python
#8.	Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len)

total_rows = len(data)

missing_values = data.isnull().sum()
missing_percentage = (missing_values / total_rows) * 100

print("Частка пропусків в кожному стовпці:")
print(missing_percentage)
```

```python
#9.	Замініть всі пропущені значення на середні значення по (стовпцю).

data_filled = data.fillna(data.mean())
data_filled
```

```python
#10.	Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату.

max_salary_year = data.loc[data['Salary in UAH'].idxmax()]['Year']
max_salary = data['Salary in UAH'].max()

print(f"Рік з найвищою зарплатою в Україні в гривнях: {max_salary_year}")
print(f"Найвища зарплата в Україні в гривнях: {max_salary}")
```

```python
#11.	Знайдіть рік з мінімальною  зарплатою в Україні в дол. США і вкажіть цю зарплату.

min_salary_year = data.loc[data['Salary in USD'].idxmin()]['Year']
min_salary = data['Salary in USD'].min()

print(f"Рік з мінімальною зарплатою в Україні в доларах США: {min_salary_year}")
print(f"Мінімальна зарплата в Україні в доларах США: {min_salary}")
```

```python
#12.	Виведіть роки, коли зарплата в грн була вище середньої.

average_salary_UAH = data['Salary in UAH'].mean()

years_above_average = data[data['Salary in UAH'] > average_salary_UAH]['Year'].tolist()

print(years_above_average)
```

```python
#13.	Порівняйте  зарплату  в грн за останні два доступних роки.

last_two_years_data = data.sort_values(by='Year', ascending=False).head(2)

salaries_UAH = last_two_years_data['Salary in UAH'].tolist()

if salaries_UAH[0] > salaries_UAH[1]:
    print(f"Зарплата в гривнях за останній рік була вище, ніж за попередній")
elif salaries_UAH[0] < salaries_UAH[1]:
    print(f"Зарплата в гривнях за останній рік була нижче, ніж за попередній")
else:
    print(f"Зарплата в гривнях за останні два роки була однаковою")
```

```python
#14.	Створіть графік, що відображає  зарплату в Україні з часом (за роками).

years = data['Year']
salaries_UAH = data['Salary in UAH']

plt.figure(figsize=(10, 6))
plt.plot(years, salaries_UAH, marker='o', linestyle='-')
plt.title('Зарплата в Україні за роками')
plt.xlabel('Рік')
plt.ylabel('Зарплата в гривнях')
plt.grid(True)

plt.show()
```

```python
#15.	Видаліть останній рік з таблиці та виведіть оновлену таблицю.

data = data.drop(data.index[-1])

print(data)
```

```python
#16.	Побудуйте графік, що відображає середню зарплату в Україні в розрізі років без останнього року.

data_without_last_year = data.drop(data.index[-1])

average_salary_by_year = data_without_last_year.groupby('Year')['Salary in UAH'].mean()

plt.figure(figsize=(10, 6))
average_salary_by_year.plot(kind='bar', color='skyblue')
plt.title('Середня зарплата в Україні за роками (без останнього року)')
plt.xlabel('Рік')
plt.ylabel('Середня зарплата в гривнях')
plt.xticks(rotation=45)

plt.show()
```

```python
#17.	Обчисліть зарплату за останні два роки і порівняйте її із зарплатою за останні 5 років. Вкажіть, чи є різниця, і в якому напрямку.

last_two_years_data = data.tail(2)

last_five_years_data = data.tail(5)

salary_last_two_years = last_two_years_data['Salary in UAH'].sum()
salary_last_five_years = last_five_years_data['Salary in UAH'].sum()

if salary_last_two_years > salary_last_five_years:
    print("Зарплата за останні два роки була вищою, ніж за останні 5 років.")
elif salary_last_two_years < salary_last_five_years:
    print("Зарплата за останні два роки була нижчою, ніж за останні 5 років.")
else:
    print("Зарплата за останні два роки була такою ж, як за останні 5 років.")
```
